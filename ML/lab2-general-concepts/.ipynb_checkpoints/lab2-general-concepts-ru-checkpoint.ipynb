```python
from sklearn import *
```

```python
columns = ['IsBadBuy', 'VehYear', 'VehicleAge', 'Make', 'Transmission', 'VehOdo', 
           'MMRAcquisitionAuctionAveragePrice',
           'MMRAcquisitionAuctionCleanPrice',
           'MMRAcquisitionRetailAveragePrice', 
           'MMRAcquisitonRetailCleanPrice',
           'MMRCurrentAuctionCleanPrice',
           'MMRCurrentAuctionCleanPrice', 
           'MMRCurrentRetailAveragePrice', 
           'MMRCurrentRetailCleanPrice'
          ]
strEncoders = []
for strColumn in ('Make', 'Transmission'):
    curEncoder = {strClass:i for i, strClass in filter(None, enumerate(data[strColumn].dropna().unique()))}
    strEncoders.append(curEncoder)

for encoder in strEncoders:
    print(encoder)

```

```python
relevant_data = data[columns].replace({'Make': strEncoders[0], 'Transmission': strEncoders[1]})
relevant_data.head(10)
```

```python
# Заменим nan на медианные значения по столбцам
relevant_data.median()
```

```python
relevant_data = relevant_data.apply(lambda x: x.fillna(x.median())) 
print(f'Nulls in data: {relevant_data.isnull().values.any()}')
```

```python
train_size = int(round(relevant_data.shape[0]*0.75))

train_ds = relevant_data.loc[0:train_size,]
test_ds = relevant_data.loc[train_size+1:relevant_data.shape[0]-1,]
```

```python
train_x = train_ds.loc[:, train_ds.columns != 'IsBadBuy']
train_y = train_ds['IsBadBuy']

test_x = test_ds.loc[:, train_ds.columns != 'IsBadBuy']
test_y = test_ds['IsBadBuy']
```

```python
model1 = linear_model.LinearRegression()
model1.fit(train_x, train_y)
```

*Функции, которые могут пригодиться при решении: `print(metrics.classification_report(...))`, `print(metrics.confusion_matrix(...))`*

Функция classification_report для каждого класса объектов считает точность (precision) в этом классе и полноту (recall). Полнота - это процент объектов данного класса, которые ваш метод предсказания тоже отнес к этому классу, среди всех объектов данного класса. Точность (precision) - то же самое, только среди всех объектов, предсказанных для этого класса.

Функция confusion_matrix возвращает матрицу с количествами объектов. Номера столбцов матрицы - это номера предсказанных классов, строки - это номера правильных классов. Например, элемент M[0,1] - это количество машин, где на самом деле IsBadBuy = 0, а вы предсказали 1.
