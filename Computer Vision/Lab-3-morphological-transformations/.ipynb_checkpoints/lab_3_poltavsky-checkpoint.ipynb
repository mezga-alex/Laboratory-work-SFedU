```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Полтавский Артем 6 группа

```python
image = cv2.imread("shesternya.png")

def show_image(image):
    plt.figure(dpi=200, figsize=(20,15))
    # plt.axis('off')
    plt.imshow(image)
show_image(image)
```

```python
def make_hole(value):
    x, y = np.indices((value * 2, value * 2))
    return (np.abs(np.hypot(value - x, value - y) - value) < 0.5).astype(np.uint8)
```

```python
hole_ring = make_hole(50)
```

```python
show_image(hole_ring)
```

# Применяемые операции

- Алгоритм морфологического расширения (dilation)
- Алгоритм морфологического сужения (erosion)
- Логические операции bitwise_or (ИЛИ), bitwise_and (И), substract (-)

```python
dots_image = cv2.erode(image, hole_ring)
show_image(dots_image)
```

```python
# элиптическое ядро MORPH_ELLIPSE
hole_mask = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (90, 90))
show_image(hole_mask)
# dilate - расширение
mask_image = cv2.dilate(dots_image, hole_mask)
show_image(mask_image)
```

```python
# к изначальному и изображению с маской

bitwise_image = cv2.bitwise_or(image, mask_image)
show_image(bitwise_image)
```

# Функция получения кольца

```python

def get_ring(bitwise_image):
    # элиптическое ядро MORPH_ELLIPSE
    gear_body = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (60, 60))
    # Сузили полное изображение
    body_image = cv2.erode(bitwise_image, gear_body)
    show_image(body_image)
    # Расширили полное изображение
    body_image_1 = cv2.dilate(body_image, gear_body)
    show_image(body_image_1)
    # Еще расширили
    ring_spacer = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7, 7))
    body_image_2 = cv2.dilate(body_image_1, ring_spacer)
    show_image(body_image_2)
    # Еще расширили
    ring_spacer_3 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (8, 8))
    body_image_3 = cv2.dilate(body_image_2, ring_spacer_3)
    show_image(body_image_3)
    # Вычли результат расширения из результата сужения
    rings_image = cv2.subtract(body_image_3, body_image_2)
    show_image(rings_image)
    return rings_image
rings_image = get_ring(bitwise_image)    
      
```

```python
# Применим логическое или с исходным изорбажением и изображением кольца чтобы оставить лишь рубцы
dotted_rings = cv2.bitwise_and(image, rings_image)
show_image(dotted_rings)
```

```python
# Расширим полученные зубцы, до момента их слияния друг с другом, но с видимым дефектом
spaced_ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (20, 20))
spaced_rings = cv2.dilate(dotted_rings, spaced_ellipse)
show_image(spaced_rings)
```

```python
# Найдем дефектные точки
def find_and_fill_holes(spaced_rings, rings_image):
    defect_cue = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (40, 40))
    # Объединим кольцо с отверстием дефекта с тонким кольцом без отверстий
    dotted_rings = cv2.bitwise_or(spaced_rings,rings_image)
    show_image(dotted_rings)
    # Вычтем из получившегося изображения кольцо с отверстием дефекта и получим дефектные точки
    dot_img = dotted_rings - spaced_rings
    show_image(dot_img)
    # Расширяем и объединяем дефектные точки с кольцом
    dots = cv2.dilate(dot_img,defect_cue)    
    result = cv2.bitwise_or(dots, spaced_rings)
    show_image(result)
    
find_and_fill_holes(spaced_rings, rings_image)
```
